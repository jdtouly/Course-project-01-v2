# D - Histogram of job satisfaction



```python
def PCTGroup(x):
    if x >= 90:
        return '90%-100%'
    elif x >= 80:
        return '80%-90%'
    elif x >= 70:
        return '70%-80%'
    elif x >= 60:
        return '60%-70%'
    elif x >= 50:
        return '50%-60%'
    elif x >= 40:
        return '40%-50%'
    else:
        return '00%-40%'
#sourcedata['Job Satisfaction (grouped)']=round(sourcedata['Job Satisfaction']/10,0)*10
sourcedata['Job Satisfaction (grouped)'] = sourcedata['Job Satisfaction'].apply(PCTGroup)

#jobsatisfaction = sourcedata['Job Satisfaction (grouped)'].groupby(['Country']).count()
jobsatisfaction = sourcedata.groupby('Job Satisfaction (grouped)')[['Country']].count()
#jobsatisfaction

graphjob = jobsatisfaction.plot.bar()
graphjob.set_title('# of countries per job satisfaction %')
```
